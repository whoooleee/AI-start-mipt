```python
import matplotlib.pyplot as plt
import time
from random import randint
```

```python
results = {
    "10": {"orel": 0, "reshka": 0},
    "100": {"orel": 0, "reshka": 0},
    "1000": {"orel": 0, "reshka": 0},
    "10000": {"orel": 0, "reshka": 0},
    "1000000": {"orel": 0, "reshka": 0}
}

def brosok(count_attemps=0):
    start = time.time()
    total_reshka = 0
    for _ in range(count_attemps):
        total_reshka += randint(0, 1)
    
    total_time = time.time() - start

    print(f"Итоговое время выполнения симуляции при {count_attemps} бросках: {total_time} секунд\nДанные по броскам: {total_reshka} выпавших решек, {count_attemps-total_reshka} орлов")
    print()

    return total_reshka / count_attemps


for attemp in results:
    new_res =brosok(int(attemp))
    results[attemp]["reshka"] = round(new_res, 3)
    results[attemp]["orel"] = round(1 - new_res, 3)

print(results)

    
counts = [int(key) for key in results]
orel_values = [results[key]["orel"] for key in results]
reshka_values = [results[key]["reshka"] for key in results]

bar_width = 0.35

fig, ax = plt.subplots()
index = range(len(counts))

plt.bar(index, orel_values, bar_width, label='Орел')
plt.bar([i + bar_width for i in index], reshka_values, bar_width, label='Решка')

plt.xlabel('Количество бросков')
plt.ylabel('Доля выпадения')
plt.title('Диаграмма выпадения орла и решки в зависимости от количества бросков')
plt.xticks([i + bar_width/2 for i in index], counts)
plt.legend()

plt.show()

    
```
